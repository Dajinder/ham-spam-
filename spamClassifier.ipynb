```python
def readData(path):
    for root, folder, files in os.walk(path):
        for filename in files:
            newPath = root +"/"+filename
            file = open(newPath, encoding='latin1')
            lines = file.readlines()
            content = []
            inBody = False
            for line in lines:
                if line == '\n':
                    inBody = True
                if inBody:
                    content.append(line)
            file.close()
            msg = "\n".join(content)
            yield msg
```

```python
def df(path,classification):
    dataset = []
    for msg in readData(path):
        row = {"message":msg,"class":classification}
        dataset.append(row)
    return pd.DataFrame(dataset)
```

```python
dataset = pd.DataFrame({"message":[],"class":[]})
dataset = dataset.append(df('emails/ham/','ham'))
dataset = dataset.append(df('emails/spam/','spam'))
```

```python
df_tokens = []
for i in range (len(dataset)):
    df_tokens.append(word_tokenize(dataset['message'].iloc[i].lower()))
```

```python
print(df_tokens[0])
```

```python
eng_stopwords = stopwords.words('english')
```

```python
eng_stopwords.extend([',','.',"'"])
```

```python
#eng_stopwords
```

```python
words = []
for list_1 in df_tokens:
    main_words = []
    for word in list_1:
        if word not in eng_stopwords:
            main_words.append(word)
    words.append(main_words)
```

```python
print(words[0])
```

```python
wnet = WordNetLemmatizer()
```

```python
for i in range(len(words)):
    for j in range(len(words[i])):
        lemm = wnet.lemmatize(words[i][j], pos='v')
        words[i][j] = lemm
```

```python
for i in range(len(words)):
    words[i] = ' '.join(words[i])
```

```python
print(words[0])
```

```python
cv = CountVectorizer()
```

```python
vect  = cv.fit_transform(words)
```

```python
vect = vect.toarray()
```
